## Law of Large Numbers in Toss of a coin

```python
import random
import matplotlib.pyplot as plt
```

```python
# Number of coin flips
n = 1000

# Initialize variables to count heads and tails
heads_count = 0
tails_count = 0

# Lists to store the proportion of heads over time
proportion_heads = []

# Simulate coin flips and calculate the proportion of heads
for i in range(1,n+1):
    # Simulate a coin flip (0 for heads, 1 for tails, assuming a fair coin)
    outcome = random.randint(0, 1)
    if outcome == 0:
        heads_count += 1
    else:
        tails_count += 1
    
    # Calculate the proportion of heads at this point
    proportion = heads_count / (heads_count + tails_count)
    proportion_heads.append(proportion)
    #print("i", i,"outcome", outcome, "heads_count",heads_count,"total count", heads_count + tails_count, "proportion heads" ,proportion_heads)
```

```python
# Plot the results
plt.plot(range(1, n + 1), proportion_heads)
plt.axhline(y=0.5, color='r', linestyle='--', label='Expected Probability (0.50)')
plt.xlabel('Number of Coin Flips')
plt.ylabel('Proportion of Heads')
plt.legend()
plt.title('Law of Large Numbers: Convergence to 0.50')
plt.show()
```
